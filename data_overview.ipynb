**3. Mối quan hệ giữa các bảng**

![Biểu đồ doanh thu](data/sodo.png)



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Giả sử bạn kiểm tra cột product_weight_g
df = pd.read_csv('D:/Ecommerce_Analysis/data/olist_order_payments_dataset.csv')
col = 'payment_value'

Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1

# Xác định ngưỡng bất thường
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Lọc các giá trị bất thường
outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
print(outliers[col].sort_values(ascending=False))
print(f"Số lượng giá trị bất thường trong cột '{col}': {len(outliers)}")



plt.figure(figsize=(10, 5))
sns.histplot(df[col], bins=50, kde=True)
plt.title(f'Phân phối giá trị của cột "{col}"')
plt.xlabel(col)
plt.ylabel('Tần suất')
plt.show()

zero_count = (df[col] == 0).sum()
print(f"Số lượng giá trị bằng 0 trong cột '{col}': {zero_count}")
```
